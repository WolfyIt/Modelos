```python
# Load libraries
import pandas as pd
from sklearn.tree import DecisionTreeClassifier # Import Decision Tree Classifier
from sklearn.model_selection import train_test_split # Import train_test_split function
from sklearn import metrics #Import scikit-learn metrics module for accuracy calculation
from sklearn.preprocessing import LabelEncoder
from sklearn.impute import SimpleImputer
import pandas as pd
```

```python
col_names = ['base_attack', 'base_defense', 'base_stamina']
# load dataset
pima = pd.read_csv("/content/drive/MyDrive/DataSets/Ejercicios/Practica 2/pokemon.csv", header=None, names=col_names)
```

```python
def fill_nan_pokemon_data(df):
    """
    Fills NaN values in the specified columns of a Pokémon dataset.

    Args:
        df (pd.DataFrame): The Pokémon dataset.

    Returns:
        pd.DataFrame: The dataset with filled NaN values.
    """

    # Specify the columns to fill NaN values for
    # Updated to only include columns that are present in the pima DataFrame
    columns_to_fill = ['base_attack', 'base_defense', 'base_stamina']

    # Fill NaN values with appropriate strategies for each column
    for column in columns_to_fill:
        if df[column].dtype == "object":  # For string columns
            df[column] = df[column].fillna("Unknown")  # Fill with "Unknown"
        elif df[column].dtype in ["int64", "float64"]:  # For numeric columns
            df[column] = df[column].fillna(df[column].mean())  # Fill with mean
        elif df[column].dtype == "bool":  # For boolean columns
            df[column] = df[column].fillna(False)  # Fill with False

    return df
```

```python
fill_nan_pokemon_data(pima)
```

```python
#split dataset in features and target variable
feature_cols = ['base_defense', 'base_stamina']

X = pima[feature_cols] # Features
y = pima['base_attack'] # Target variable

X = X.iloc[1:]  # Esto elimina la primera fila
```

```python
y = y.iloc[1:]  # Esto elimina la primera fila
```

```python
# Split dataset into training set and test set
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=1) # 70% training and 30% test
```

```python
y_train
```

```python
for column in X_train.columns:
    if X_train[column].dtype == 'object':
        # Convert boolean strings to actual booleans
        X_train[column] = X_train[column].replace({'True': True, 'False': False})
        # Apply Label Encoding to the column if it is categorical
        le = LabelEncoder()
        X_train[column] = le.fit_transform(X_train[column].astype(str))

for column in X_test.columns:
    if X_test[column].dtype == 'object':
        # Convert boolean strings to actual booleans
        X_test[column] = X_test[column].replace({'True': True, 'False': False})
        # Apply Label Encoding to the column if it is categorical
        le = LabelEncoder()
        X_test[column] = le.fit_transform(X_test[column].astype(str))
```

```python
# Create Decision Tree classifer object
clf = DecisionTreeClassifier()

# Train Decision Tree Classifer
clf = clf.fit(X_train,y_train)

#Predict the response for test dataset
y_pred = clf.predict(X_test)
```

```python
y_pred
```

```python
# Model Accuracy, how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test, y_pred))
```
